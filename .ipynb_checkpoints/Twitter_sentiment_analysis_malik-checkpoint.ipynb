## Text pre-processing

```python
# Text cleaning and pre-processing 
#Removeing every symbol except alphanumeric , transforming all words to lower case , and remove stopwords 
stop_words = stopwords.words('english')
data['clean_text'] = data['clean_text'].apply(lambda x:' '.join(x for x in x.split() if x not in stop_words))
clean_corpus = data['clean_text']
corpus = []
for text in clean_corpus:
    text = re.sub('[^a-zA-Z]',' ', text)
    text = text.lower()
    test = text.split()
    text = ''.join(text)
    corpus.append(text)
# Transformed data    
clean_corpus_data = corpus
```

```python
# Appending the transformed data in dataset
data['clean_corpus'] = clean_corpus_data
# Droping the clean_text column
data = data.drop('clean_text', axis = 1)
```

```python
# Converting data to list 
corpus_2 = data['clean_corpus'].tolist()
corpus_2
```

## OneHot Representation 

```python
 # Vocabulary size (setting the vocabulary size)
 voc_size = 50000
```

```python
# OneHot Representation for independent variabel
onehot_rep = [one_hot(words,voc_size)for words in corpus_2]
onehot_rep
```

```python
# Observing the 103 index sentence
onehot_rep[103]
```

## Embedded Layer

```python
# Embedded layer creation
# Adding the pad sequence (pre padded layer )
sent_length = 30
embedded_doc = pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
# Length of embedded doc
len(embedded_doc)
```

```python
# Observing the 0 index embedded doc
embedded_doc[0]
```

# MODEL BUILDING
we will build two model 
1.   Machine lerning model -- BernoulliNB Model
2.   Deep learning model -- LSTM Model






## Bernoulli Naive Bayes Model

```python
# setting the feature(input) and lable(output) for BernoulliNB Model
xnb = embedded_doc
ynb = data['category']
```

```python
# shape of x and y
xnb.shape , ynb.shape
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(xnb,ynb,test_size=0.3,random_state=100)
```

```python
# BernoulliNB Model
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
nb_model = BernoulliNB()
nb_model.fit(x1_train, y1_train)
nb_model_pred = nb_model.predict(x1_test)
nb_score = nb_model.score(x1_test,y1_test)
print("Naive Bayes score: ",nb_score)
```

```python
# Classification report
print(classification_report(y1_test,nb_model_pred))
```

```python
# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(nb_model, x1_test, y1_test,cmap='Blues')
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

# LSTM Neural Network Model

```python
# LSTM Model
embedding_vector_features = 200
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
# input and output variable
x = np.array(embedded_doc)
# converting categorical variables to numerical variables for lable(y)-output
y = pd.get_dummies(data['category']).values
print('Shape of Label tensor: ', y.shape)
```

```python
# Train test split
x2_train,x2_test,y2_train,y2_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
# LSTM Neural Network Model Training
history = model.fit(x2_train,y2_train,validation_data=(x2_test,y2_test),epochs=10,batch_size=64)
```

```python
# plotting the accuracy graph
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# LSTM Model Accuracy
model_score = model.evaluate(x2_test,y2_test)
print("LSTM MODEL ACCURACY : ",model_score)
```

The LSTM model is overfitting --(Low Bias and High Variance)
we need the generalized model --(Low Bias and Low Variance)

-----hence we will add dropout layer and early stopping----- 


```python
# Adding the Dropout layer and early stopping to reduce the variance
tf.keras.backend.clear_session()

embedding_vector_features = 100
drop_model=Sequential()
drop_model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
drop_model.add(Dropout(0.3))
drop_model.add(LSTM(100))
drop_model.add(Dropout(0.3))
drop_model.add(Dense(3,activation='softmax'))
drop_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(drop_model.summary())
```

```python
# Early stopping 
from keras import callbacks
earlystopping = callbacks.EarlyStopping(monitor = 'val_loss',mode = 'min',patience = 5,
                                        restore_best_weights = True)
```

```python
# Train test split
x3_train,x3_test,y3_train,y3_test = train_test_split(x,y,test_size=0.3,random_state=50,shuffle=True)
```

```python
# Training the model with dropout layer and early stopping
history2 = drop_model.fit(x3_train,y3_train,validation_data=(x3_test,y3_test),epochs=30,batch_size=32,callbacks = [earlystopping])
```

```python
# ploting accuracy graph
plt.plot(history2.history['accuracy'])
plt.plot(history2.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Model accuracy
drop_model_score = drop_model.evaluate(x3_test,y3_test)
print("LSTM MODEL ACCURACY : ",drop_model_score)
accuracy = drop_model_score[1]*100

print('Test Accuracy: ', drop_model_score[1]*100, '%')
```

```python
# Prediction on test data
predicted = drop_model.predict(x3_test)
drop_model_pred = predicted.argmax(axis=-1)
```

```python
# Model prediction
drop_model_pred
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y3_test.argmax(axis=-1),drop_model_pred))
```

```python
# Confusion Matrix
yy = y1_test
cm = confusion_matrix(y3_test.argmax(axis=-1), drop_model_pred)
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap=plt.cm.Blues, cbar=False)
ax.set(xticklabels=list(yy.unique()), yticklabels=list(yy.unique()), title="Confusion matrix")
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.set_ylabel("True", color="royalblue", fontsize=35, fontweight=700)
ax.set_xlabel("Prediction", color="royalblue", fontsize=35, fontweight=700)
plt.yticks(rotation=0)
plt.show()
```
